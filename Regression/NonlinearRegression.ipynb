```python
import pandas as pd
df = pd.read_csv('china.csv')
print(df.head())
print(df.tail())
```

```python
x = (np.arange(1960, 2015, 1))/2014
def sigmoid(x, alpha, beta):
    y = 1.0/(1.0 + np.exp(-alpha*(x - beta)))
    return y
```

```python
xdata = df.Year.values
ydata = df.Value.values

X = xdata/np.max(xdata)
Y = ydata/np.max(ydata)
```

```python
from scipy.optimize import curve_fit
popt, pcov = curve_fit(sigmoid, X, Y)
print(popt)
```
